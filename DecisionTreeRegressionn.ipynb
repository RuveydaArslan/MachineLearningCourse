```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df["fuel"].unique()
```

```python
df = pd.get_dummies(df, columns=["fuel"])
```

```python
df.head()
```

```python
df["fuel_CNG"] = df["fuel_CNG"].astype(int)
df["fuel_Diesel"] = df["fuel_Diesel"].astype(int)
df["fuel_Electric"] = df["fuel_Electric"].astype(int)
df["fuel_LPG"] = df["fuel_LPG"].astype(int)
df["fuel_Petrol"] = df["fuel_Petrol"].astype(int)
```

```python
df.info()
```

```python
df["seller_type"].unique()
```

```python
df = pd.get_dummies(df,columns=["seller_type"])
```

```python
df.head()
```

```python
df["seller_type_Dealer"] = df["seller_type_Dealer"].astype(int)
df["seller_type_Individual"] = df["seller_type_Individual"].astype(int)
df["seller_type_Trustmark Dealer"] = df["seller_type_Trustmark Dealer"].astype(int)
```

```python
df.head()
```

```python
df["transmission"].unique()
```

```python
df = pd.get_dummies(df,columns=["transmission"])
```

```python
df["transmission_Automatic"] = df["transmission_Automatic"].astype(int)
df["transmission_Manual"] = df["transmission_Manual"].astype(int)
```

```python
df.info()
```

```python
df["owner"].unique()
```

```python
df = pd.get_dummies(df,columns=["owner"])
```

```python
df.head()
```

```python
df["owner_First Owner"] = df["owner_First Owner"].astype(int)
df["owner_Fourth & Above Owner"] = df["owner_Fourth & Above Owner"].astype(int)
df["owner_Second Owner"] = df["owner_Second Owner"].astype(int)
df["owner_Test Drive Car"] = df["owner_Test Drive Car"].astype(int)
df["owner_Third Owner"] = df["owner_Third Owner"].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
df.drop("name",axis=1,inplace=True)
```

```python
X = df.drop("selling_price",axis=1)
y = df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=15)
```

```python
X_train.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_model = DecisionTreeRegressor(criterion="absolute_error",max_depth=3,random_state=0)
```

```python
X_train.select_dtypes(include="object").columns
```

```python
df.info()
```

```python
tree_model.fit(X_train,y_train)
```

```python
y_pred = tree_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
print("Mean Absolute Error: ",mean_absolute_error(y_test,y_pred))
print("Mean Squared Error: ",mean_squared_error(y_test,y_pred))
print("R2 Score: ",r2_score(y_test,y_pred))
plt.scatter(y_test,y_pred)
plt.show()
```
